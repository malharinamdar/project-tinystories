```python

endings = ["happy", "sad", "mysterious","inspiring"]  # Your tone options
```

```python
# Prompt template
prompt_template = (
    "Write a simple Hindi children's story suitable for a 5-year-old like Panchtantra tales, "
    "using easy and understandable language (under 200 words). The story should have a clear structure "
    "(don't give any title). Use the verb '{verb}', the noun '{noun}', and the adjective '{adjective}' "
    "naturally within the story. Ensure that subject-verb agreement, tense consistency, and simple sentence structure "
    "are followed. The story should have at most two dialogues. The story must naturally lead to a {ending} ending. "
    "Do not use the word {ending} directly in the story. Instead, convey these emotions through the resolution of the story, "
    "showing characters’ reactions, the atmosphere, and the conclusion. Make sure the ending reflects the chosen tone. "
    "Give the moral of the story at last without any heading. The moral should align with the tone of the story and the resolution."
)
```

```python
def generate_story(noun, verb, adjective, ending):
    # Fill the prompt with the selected word and tone
    prompt = prompt_template.format(noun=noun, verb=verb, adjective=adjective, ending=ending)
    
    # Call the GPT API to generate a response
    response = client.chat.completions.create(
        model="claude-3.5-sonnet",
        messages=[{"role": "user", "content": prompt}],
    )
    story = response.choices[0].message.content
    return story
```

```python
def create_dataset(nouns, verbs, adjectives, endings, num_stories=10000):
    dataset = []
    for _ in range(num_stories):
        noun = random.choice(nouns)
        verb = random.choice(verbs)
        adjective = random.choice(adjectives)
        ending = random.choice(endings)
        try:
            story = generate_story(noun, verb, adjective, ending)
            dataset.append({
                "noun": noun,
                "verb": verb,
                "adjective": adjective,
                "ending": ending,
                "story": story
            })
        except Exception as e:
            print(f"Error generating story: {e}")
    
    return dataset
```

```python
def save_dataset_to_json(dataset, filename="hindi_stories_dataset112.json"):
    with open(filename, "w", encoding="utf-8") as f:
        json.dump(dataset, f, ensure_ascii=False, indent=4)

```

```python
# Main execution
if __name__ == "__main__":
    num_stories = 50  # Adjust the number of stories as needed
    print(f"Generating {num_stories} stories...")
    dataset = create_dataset(nouns, verbs, adjectives, endings, num_stories=num_stories)
    print("Saving dataset...")
    save_dataset_to_json(dataset)
    print(f"Dataset saved successfully with {len(dataset)} stories!")
```
