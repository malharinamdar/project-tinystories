```python
def send_requests(results):
    file_lock = threading.Lock()
    # Initialize CSV with headers
    with open(csv_path, 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(["story", "noun", "verb", "adjective", "end"])
    for i in range(100):  # Adjust the number of iterations as needed
        try:
            print(f"Sending request for story {i}")
            # Randomly sample words and endings
            ending = random.choice(endings)
            verb = random.choice(verbs)
            noun = random.choice(nouns)
            adjective = random.choice(adjectives)
            
            # Hindi-specific prompt
            basePromptHindi = f'''Write a simple Hindi children's story suitable for a 5-year-old like Panchtantra tales, 
            using easy and understandable language (under 200 words). The story should have a clear structure (don't give any title). 
            Use the verb "{verb}", the noun "{noun}", and the adjective "{adjective}" naturally within the story. 
            Ensure that subject-verb agreement, tense consistency, and simple sentence structure are followed. 
            The story should include at most two dialogues between the characters, keeping them simple and clear. 
            Avoid giving any specific names to characters; refer to them by their roles or descriptions. 
            The story must naturally lead to a {ending} ending. The interactions and actions should flow naturally to ensure the story makes sense. 
            Do not use the word {ending} directly in the story. Incorporate natural elements like trees, rivers, animals, or village life to make the story more engaging for children. 
            Introduce supporting characters like animals, villagers, or magical creatures to add depth. Make sure the ending reflects the chosen tone. 
            Without any heading, give the moral of the story at last. The moral should align with the tone of the story and the resolution.'''
            
            # Send the request
            response = client.chat.completions.create(
                model="gpt-4o-mini",
                messages=[{"role": "assistant", "content": basePromptHindi}],
            )
            story = response.choices[0].message.content
            print(f"Response received for story {i}")
            
            # Save the story and metadata
            new_row = {
                "story": story,
                "noun": noun,
                "verb": verb,
                "adjective": adjective,
                "end": ending
            }
            with file_lock:
                with open(csv_path, 'a', newline='', encoding='utf-8') as file:
                    writer = csv.writer(file)
                    writer.writerow([
                        story, 
                        noun, 
                        verb, 
                        adjective, 
                        ending
                    ])
            results.append(new_row)
        except Exception as e:
            logging.error(f"Error occurred: {e}")

```

```python
def create_parallel_requests(n=15):
    threads = []
    results = []  # Shared results list
    for i in range(n):  # Number of threads
        thread = threading.Thread(target=send_requests, args=(results,))
        threads.append(thread)
        thread.start()
    for thread in threads:
        thread.join()
    return results

```
